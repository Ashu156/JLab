```python
import numpy as np
import pandas as pd
import io
```

```python
from google.colab import files
uploaded = files.upload()
df1 = pd.read_excel(io.BytesIO(uploaded['Social linear rewards.xlsx']), sheet_name = 'Sheet1')
df2 = pd.read_excel(io.BytesIO(uploaded['Social linear rewards.xlsx']), sheet_name = 'Sheet2')
```

```python
import matplotlib.pyplot as plt

# plt.style.use("dark_background")
for param in ['text.color', 'axes.labelcolor', 'xtick.color', 'ytick.color']:
    plt.rcParams[param] = '0.9'  # very light grey
for param in ['figure.facecolor', 'axes.facecolor', 'savefig.facecolor']:
    plt.rcParams[param] = '#212946'  # bluish dark grey
colors = [
    # '#08F7FE',  # teal/cyan
    '#FE53BB',  # pink
    # '#F5D300',  # yellow
    '#00ff41',  # matrix green
]
fig, ax = plt.subplots()
df1.plot(marker='o', color=colors, ax=ax)

ax.grid(color='#2A3459')
ax.set_xlim([ax.get_xlim()[0], ax.get_xlim()[1]])  # to not have the markers cut off
ax.set_ylim([0.0, 7.0])
ax.set_xlabel('Training session #')
ax.set_ylabel('Rewards /minute'),
# ax.set_xticklabels([])
ax.axhline(4.0, linestyle='--', color = 'red')
# ax.axvline(3.5, linestyle='--', color = 'white')
```
